##Sobrecarga de operadores

Ejercicio en python de **overloading**, primero se hara el ejercicio sin overloading y luego con overloading.

Haremos una clase Vector en la cual vamos a representar sus operaciones.



```python
#Creamos la clase Vector y sus metodos.
class Vector(object):
  def __init__(self,x,y):
    self.x = x
    self.y = y

  def __repr__(self):
    return '<Vector (%f, %f)>' % (self.x, self.y)
```

```python
#Instanciamos dos objetos de tipo Vector.
a = Vector(1,2)
b = Vector(3,4)

a
```

##Operaciones sin sobrecarga

###Suma
La Suma de los vectores es la suma de cada uno de los componentes x,y.

```python
a = Vector(2,3)
b = Vector(3,7)

x1 = a.x + b.x
y1 = a.y + b.y

result = Vector(x1, y1)
print(result)
```

###Resta
La resta de dos vectores es la resta de cada uno de sus componentes x,y.


```python
a = Vector(5,8)
b = Vector(3,2)

x1 = a.x - b.x
y1 = a.y - b.y

result = Vector(x1, y1)
print(result)
```

###Multiplicacion por un escalar

Se realiza multiplicando cada uno de los componentes x,y por un escalar.

```python
a = Vector(3,5)
c = 1.654

x1 = a.x * c
y1 = a.y * c

result = Vector(x1, y1)
print(result)
```

##Operaciones con sobrecarga de operadores



```python
class Vector(object):
  def __init__(self,x,y):
    self.x = x
    self.y = y

  def __add__(self, v):
    #Sumar dos vectores
    return Vector(self.x + v.x, self.y + v.y)
  
  def __sub__(self, v):
    #Restar dos vectores
    return Vector(self.x - v.x, self.y - v.y)
  
  def __mul__(self, s):
    #Multiplicar un vector por un escalar
    return Vector(self.x * s, self.y * s)

  def __repr__(self):
    return '<Vector (%f, %f)>' % (self.x, self.y)
```

```python
#Sumando dos vectores con overloading:

a = Vector(3, 5)
b = Vector(4, 5)

print(a + b)
```

```python
#Restando dos vectores con overloading:

a = Vector(8, 10)
b = Vector(4, 5)

print(a - b)
```

```python
#Multiplicacion de vectores por un escalar con overloading:

a = Vector(8, 10)
b = Vector(4, 5)

print(b * 1.3)
```
